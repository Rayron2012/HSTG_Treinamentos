#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
$C = \frac{5}{9}(F-32)$

```python
farenheit = float(input('Digite a temperatura em Farenheit: '))
celcius = ((farenheit - 32) / 1.8)

print('A temperatura em Celcius é de: {:.3}°C'.format(celcius))
```

#### 9. Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
$F = \frac{9}{5}C + 32$

```python
celcius = float(input('Digite seus graus Celcius: '))
farenheit = (celcius * 1.8) + 32

print('Aconversão de Celcius para Farenheit é de: {:.3}°F'.format(farenheit))
```

#### 10. Tendo como dados de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, usando a seguinte fórmula:

#### 11. Tendo como dado de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, utilizando as seguintes fórmulas:
##### a. Para homens: $P = 72,7h - 58$
##### b. Para mulheres: $P = 62,1h - 44,7$

```python
sexo = input('Digite seu sexo (M/F): ').upper()

if sexo == 'M':
    alturam = float(input('Digite sua altura em metros: '))
    peso_ideal_m = (alturam * 72.7) - 58
    print('Seu peso ideal é de: {:.2f}kg'.format(peso_ideal_m))
elif sexo == 'F':
    alturaf = float(input('Digite sua altura em metros: '))
    peso_ideal_f = (alturaf * 62.1) - 44.7
    print('Seu peso ideal é de: {:.2f}kg'.format(peso_ideal_f))
else:
    print('Sexo inválido. Por favor, digite M para masculino ou F para feminino')
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
salario = int(input('Qual é o salário do funcionário? R$ '))

ip = 11 / 100

print('Do sallario de R$ {}, o imposto pago é de R$ {:.2f}'.format(salario, salario * ip))
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
print('Do seu (salario de R$ {}, o inss descontado é de {}'.format(salario, salario * 0.08))
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
print('Do seu salario de {}, vc paga de sindicato R$ {:.2f}'.format(salario, salario * 0.05))
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
descontos = (salario * 0.11) + (salario * 0.08) + (salario * 0.05)
salario_liquido = salario - descontos

print('Seu salario liquido é de R$ {}'.format(salario_liquido))
```

#### 13. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
pintada = int(input('Quantos metros quadrados serão pintados? '))
litros_necessario = pintada / 3
latas_necessarias = litros_necessario / 18
custo_por_lata = 80
total_custo = latatas_necessarias * custo_por_lata

print('Para pintar {}m², serão necessárias {} latas de tinta, custando um total de {:3f}.'.format(pintada, latas_necessarias, total_custo))
```

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
tamanho = float(input('Digite o tamanho do arquivo (MB): '))
vel = float(input('Digite a velocidade da Internet (Mbps): '))
tamanho_megabits = tamanho * 8
tempo_segundos = tamanho_megabits / vel
tempo_minutos = tempo_segundos / 60


print('Voce levara aproximadamente: {:.2f} Minutos, te tempo para baixar seu arquivo'.format(tempo_minutos))
```
