### F-string

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo

# Com format
print('O faturamento das sua vendas é de: {}. O custo foi de {}. Sobrando um lucro de {}'.format(faturamento, custo, lucro))

# Com f-String
print(f'O faturamento das sua vendas é de: {faturamento}. O custo foi de {custo}. Sobrando um lucro de {lucro}')
```

### Não tem diferença entre eles, vai do gosto de cada um
