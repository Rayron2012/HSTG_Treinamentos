#### Na maioria das vezes vasmos precisar tipificar as entradas dos usuarios, e por isso que é importante tipificar ou mudar determinados tipos de variiaveis

```python
faturamento = input('Insira o valor faturado')
custo = input('Insira o custo')

print(type(faturamento))
print(type(custo))

lucro = faturamento - custo
print(lucro)
```

```python
faturamento = int(input('Insira o valor faturado'))
custo = int(input('Insira o custo'))

print(type(faturamento))
print(type(custo))

lucro = faturamento - custo
print(lucro)
```

##### Podemos usar os tipos mais comuns que são: str, int e float
