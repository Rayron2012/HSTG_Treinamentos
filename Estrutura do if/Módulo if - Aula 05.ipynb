# Casos com várias condições/comparações

### Estrutura:

Quando temos várias comparações, ao invés de criar if dentro de if podemos usar os operadores "and" e "or" para tratar essas condições.

Funciona assim:

### Exemplo

Vamos voltar ao exemplo de cálculo de meta de vendas dos funcionários. Muitas empresas atribuem bonificação do salário dos funcionários de acordo com o resultado do funcionário e também com o resultado da empresa como um todo.

Nesse caso, a regra funciona da seguinte forma:
- Se o funcionário vendeu mais do que a meta de vendas e a loja bateu a meta de vendas da loja, o funcionário ganha 3% do que ele vendeu em forma de bônus.
- Caso o funcionário tenha batido a meta de vendas individual dele, mas a loja não tenha batido a meta de vendas da loja como um todo, o funcionário não ganha bônus.

```python
venda_funcionario = float(input('Quando voce vendeu esse mes? R$ '))
venda_loja = float(input('Quanto a loja vendeu esse mes? R$ '))
meta_loja= 20000
meta_funcionario = 5000
bonus = 0

if venda_loja >= meta_loja and venda_funcionario >= meta_funcionario:
    bonus = venda_funcionario * 0.03
    print('Parabens! Voce atingiu a meta e ganhou um bonus de R$ {:.2f}'.format(bonus))

elif venda_funcionario < meta_funcionario and venda_loja < meta_loja:
    print('Nem voce nem a loja atingiram a meta. Nao ha bonus esse mes.')

elif venda_funcionario >= meta_funcionario or venda_loja < meta_loja:
    print('Voce atingiu a meta individual, mas a loja nao atingiu a meta. Nao ha bonus esse mes.')

else:
    print(f"Infelizmente voce não ganhou um bonus esse mes pois ficou R$ {venda_funcionario - meta_funcionario} abaixo da meta.")
```

### Outro exemplo

Agora vamos levar essa análise mais a fundo.

Nessa empresa, existe um outro caso também que garante que o funcionário ganhe um bônus, independente das vendas que ele fez naquele mês.

Todo mês os diretores da empresa fazem uma avaliação qualitativa de todos os funcionários. Nessa avaliação os diretores dão uma nota de 0 a 10 para cada funcionário. Se a nota do funcionário for 9 ou 10, ele também ganha o bônus de 3% do valor de vendas. (os bônus não são cumulativos)

```python
nota_f = int(input('Qual a sua nota de satisfacao esse mes? (0 a 10) '))
meta_nota = 9

if venda_funcionario < meta_funcionario and nota_f >= meta_nota:
    bonus = venda_funcionario * 0.03
    print('Parabens! voce náo bateu a meta de vendas, mas ganhou um bonus graças a sua nota de avaliação que foi de {}! Seu bonus é de R$ {:.2f}'.format(nota_f, bonus))

elif venda_funcionario >= meta_funcionario and nota_f < meta_nota:
    bonus = venda_funcionario * 0.03
    print(f'Voce bateu a meta de vendas, mas nao atingiu a meta de nota de satisfacao. bonus em cima de vendas de {bonus}.')

elif venda_funcionario < meta_funcionario and venda_loja < meta_loja and nota_f < meta_nota:
    print('Infelizmente voce nao ganhou bonus esse mes pois nao bateu nenhuma das metas. Nem a loja')

elif venda_loja < meta_loja and nota_f > meta_nota:
    bonus = venda_funcionario * 0.03
    print('Parabens! A loja nao bateu a meta de vendas, mas voce ganhou um bonus graças a sua nota de avaliação que foi de {}! Seu bonus é de R$ {:.2f}'.format(nota_f, bonus))

else:
    bonus = venda_funcionario * 0.03
    print('Parabéns voce bateu ambas as metas, porem o total de bonus que voce pode receber é R$ {:.2f}'.format(bonus))
```
