#### 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

````python
user = input("Digite um nome de usuário: ")
passw = input("Digite uma senha: ")

while user == passw:
    print("Erro! Usuário e senha não podem ser iguais.")
    user = input("Digite um nome de usuário: ")
    passw = input("Digite uma senha: ")

print("Cadastro realizado com sucesso!")
````

#### 3. Faça um programa que leia e valide as seguintes informações (e para cada uma delas, continue pedindo a informação até o usuário inserir corretamente):
##### Nome: maior que 3 caracteres;
##### Idade: entre 0 e 150;
##### Salário: maior que zero;
##### Sexo: 'f' ou 'm';
##### Estado Civil: 's', 'c', 'v', 'd';

````python
# NOME (mais de 3 caracteres)
while True:
    nome = input("Digite seu nome: ").strip()
    if len(nome) > 3:
        break
    print("Nome inválido! Deve ter mais de 3 caracteres.")

# IDADE (entre 0 e 150) + validar se é número
while True:
    idade_txt = input("Digite sua idade: ").strip()
    if not idade_txt.isdigit():
        print("Idade inválida! Digite um número inteiro.")
        continue

    idade = int(idade_txt)
    if 0 <= idade <= 150:
        break
    print("Idade inválida! Deve estar entre 0 e 150.")

# SALÁRIO (maior que zero) + validar se é número (aceita decimal)
while True:
    salario_txt = input("Digite seu salário: ").strip().replace(",", ".")
    try:
        salario = float(salario_txt)
    except ValueError:
        print("Salário inválido! Digite um número (ex: 1500 ou 1500.50).")
        continue

    if salario > 0:
        break
    print("Salário inválido! Deve ser maior que zero.")

# SEXO ('f' ou 'm')
while True:
    sexo = input("Digite seu sexo (F/M): ").strip().lower()
    if sexo in ("f", "m"):
        break
    print("Sexo inválido! Use 'f' ou 'm'.")

# ESTADO CIVIL ('s', 'c', 'v', 'd')
while True:
    estado_civil = input("Digite seu estado civil (S/C/V/D): ").strip().lower()
    if estado_civil in ("s", "c", "v", "d"):
        break
    print("Estado civil inválido! Use 's', 'c', 'v' ou 'd'.")

print("\nCadastro concluído com sucesso!")
print(f"Nome: {nome} | Idade: {idade} | Salário: {salario:.2f} | Sexo: {sexo} | Estado civil: {estado_civil}")
````

Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

````python
pais_a = 80000
pais_b = 200000
taxa_a = 0.03
taxa_b = 0.015 

anos = 0
while pais_a < pais_b:
    pais_a += pais_a * taxa_a
    pais_b += pais_b * taxa_b
    anos += 1
print(f"Serão necessários {anos} anos para que a população do país A ultrapasse a população do país B.")    
````

#### 5. Altere o programa anterior permitindo ao usuário informar as populações e as taxas de crescimento iniciais. Valide a entrada e permita repetir a operação.

````python
while True:
    # entrada das populações
    try:
        pais_a = int(input("População inicial do país A: "))
        pais_b = int(input("População inicial do país B: "))

        if pais_a <= 0 or pais_b <= 0:
            print("As populações devem ser maiores que zero.")
            continue

        # entrada das taxas (em %)
        taxa_a = float(input("Taxa de crescimento do país A (%): ")) / 100
        taxa_b = float(input("Taxa de crescimento do país B (%): ")) / 100

        if taxa_a <= 0 or taxa_b <= 0:
            print("As taxas devem ser maiores que zero.")
            continue

    except ValueError:
        print("Entrada inválida. Use apenas números.")
        continue

    # cálculo
    anos = 0
    while pais_a < pais_b:
        pais_a += pais_a * taxa_a
        pais_b += pais_b * taxa_b
        anos += 1

    print(f"Serão necessários {anos} anos para o país A ultrapassar ou igualar o país B.")

    # repetir?
    repetir = input("Deseja repetir o cálculo? (s/n): ").lower()
    if repetir != "s":
        break
````

#### 6. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o maior faturamento

````python
meses = 0
fat = []

while True:
    faturamento = int(input("Digite o faturamento mensal da empresa: "))
    fat.append(faturamento)
    meses += 1
    if meses >= 5:
        break

print(f"O maior faturamento mensal da empresa é de: R$ {max(fat):.2f}")
````

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).

````python
meses = 0
fat = []

while True:
    faturamento = int(input("Digite o faturamento mensal da empresa: "))
    fat.append(faturamento)
    meses += 1
    if meses >= 5:
        break

print(f"O faturamento mensal da empresa é: R$ {sum(fat):.2f}, e a média mensal é: R$ {sum(fat)/meses:.2f}")
````

#### 8. Faça um programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa); e então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

````python
idades = []

while True:
    texto = input("Digite a idade da pessoa (0 para sair): ").strip()

    if texto == "":
        print("Digite um número, não deixa vazio.")
        continue

    if not texto.isdigit():
        print("Só vale número inteiro (ex: 23).")
        continue

    idade_atual = int(texto)

    if idade_atual == 0:
        break

    idades.append(idade_atual)

if len(idades) == 0:
    print("Nenhuma idade foi informada.")
else:
    media = sum(idades) / len(idades)

    if media <= 25:
        print(f"A média é {media:.0f} anos. Jovem.")
    elif media <= 60:
        print(f"A média é {media:.0f} anos. Sênior.")
    else:
        print(f"A média é {media:.0f} anos. Idosa.")
````

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

````python
candidatos = ["A", "B", "CX"]
eleitores = []

qtd_eleitor = int(input("Digite a quantidade de eleitores: "))
for i in range(qtd_eleitor):
    voto = input(f"Eleitor {i+1}, digite seu voto (1 = A, 2 = B, 3 = CX): ").strip().upper()
    while voto not in ["1", "2", "3"]:
        print("Voto inválido! Tente novamente.")
        voto = input(f"Eleitor {i+1}, digite seu voto (1 = A, 2 = B, 3 = CX): ").strip()
    if voto == "1":
        eleitores.append("A")
    elif voto == "2":
        eleitores.append("B")
    else:
        eleitores.append("CX")
print("\nResultado da eleição: ")
for candidato in candidatos:
    votos = eleitores.count(candidato)
    print(f"Candidato {candidato}: {votos} votos {eleitores}")
````
