```python
produtos = ["iphone", "ipad", "airpod", "macbook"]
precos = [7000, 10000, 2500, 14000]

```

### For item in lista

```python
# preco com imposto
for preco in precos:
    print(round(preco * 1.1))


# Melhor forma de usar quando queremos apenas um iterable, por padrao é melhor usar for item lista
```

### For i in range

```python
# preco de cada produto

for i in range(len(produtos)):
    print("O preço do {} é de R${}\n".format(produtos[i], precos[i]))


# Mas caso exista mais de uma lista a melhor opcao é usar o for in range. 
```

### For item in lista com enumerate

```python
# preco de cada produto com imposto

for i, preco in enumerate(precos):
    preco = round(preco * 1.1)
    produto = produtos[i]
    print(produto, preco)


# Vocé pode usar o enumerate, quando precisar pegar ou controlar o indice é quase a mesma coisa do acima porem um pouco mais pesado
```
