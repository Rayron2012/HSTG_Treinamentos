## 1. A Ordem das Coisas: O Ciclo de Vida de uma Function
**Onde aplicar:** Quando vocÃª precisa repetir uma tarefa (ex: limpar um log, validar um IP).

### A Regra de Ouro:
1. **DefiniÃ§Ã£o (`def`):** VocÃª cria a ferramenta.
2. **Chamada:** VocÃª usa a ferramenta.
3. **Modularidade:** Uma funÃ§Ã£o pode chamar outra para dividir tarefas complexas.

```python
# Exemplo: Uma funÃ§Ã£o chamando outra (Modularidade)

def limpar_texto(texto):
    return texto.strip().lower()

def processar_log(entrada):
    # Aqui a funÃ§Ã£o 'processar_log' utiliza a 'limpar_texto'
    texto_limpo = limpar_texto(entrada)
    return f"[LOG PROCESSADO]: {texto_limpo}"

print(processar_log("  ERRO NO SERVIDOR DE BANCO  "))
```

## 2. Desvendando `*args` e `**kwargs`
**Onde aplicar:** Quando vocÃª nÃ£o sabe quantos itens vai receber (ex: uma lista de IPs para bloquear ou vÃ¡rios dados de um usuÃ¡rio).

* `*args`: Recebe vÃ¡rios valores e os guarda em uma **Tupla** (lista imutÃ¡vel).
* `**kwargs`: Recebe vÃ¡rios valores com nome (chave=valor) e guarda em um **DicionÃ¡rio**.

```python
# Exemplo PrÃ¡tico

def registrar_acesso(*ips, **detalhes):
    print(f"Bloqueando os IPs: {ips}")
    if 'motivo' in detalhes:
        print(f"Motivo do bloqueio: {detalhes['motivo']}, UsuÃ¡rio: {detalhes.get('usuario', 'Desconhecido')}")

registrar_acesso("192.168.0.1", "10.0.0.5", motivo="Brute Force", usuario="admin")
```

### ðŸ“ ExercÃ­cio 2: O Coletor de Dados
Crie uma funÃ§Ã£o `cadastrar_ferramenta` que receba o nome da ferramenta como primeiro argumento fixo, vÃ¡rios `*args` para tags de categorias (ex: 'seguranÃ§a', 'python') e `**kwargs` para especificaÃ§Ãµes tÃ©cnicas (ex: versao='1.0').

```python
# Implemente aqui
def cadastrar_ferramenta(nome_ferramenta, *ferramenta, **versao):
    listagem = []
    for programa in ferramenta:
        listagem.append(f"{nome_ferramenta} - {programa} - VersÃ£o: {versao.get('versao', 'Desconhecida')}")
    return listagem

ferramentas = cadastrar_ferramenta("AntivÃ­rus", "Norton", "McAfee", versao="2024.1")
ips = cadastrar_ferramenta("IPS", "Snort", "Suricata", versao="3.0")

print(ferramentas,"\n",ips)
```

## 3. List Comprehensions (O Jeito Pythonico)
**Onde aplicar:** Para transformar listas de forma rÃ¡pida sem precisar de 4 linhas de `for` e `append`.

**FÃ³rmula:** `[resultado for item in lista if condicao]`

```python
# Jeito antigo:
numeros = [1, 2, 3, 4, 5]
dobro = []
for n in numeros:
    dobro.append(n * 2)

# Com List Comprehension:
dobro_pythonico = [n * 2 for n in numeros]
print(dobro_pythonico)
```

### ðŸ“ ExercÃ­cio 3: Filtro de Vulnerabilidades
Dada a lista `severidades = [1, 8, 3, 10, 5, 9]`, use List Comprehension para criar uma nova lista apenas com as severidades maiores ou iguais a 8.

```python
severidades = [1, 8, 3, 10, 5, 9]
# Implemente aqui

ligar_p_cliente = [n for n in severidades if n >= 8]
print(ligar_p_cliente)
```
