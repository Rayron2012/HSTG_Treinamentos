- Exemplos de Lista:

Listas de Produtos de uma Loja:

#### No python as listas foram pensadas para serem homogenias, ou seja, somente texto ou somente numeros, evitando de misturar numeros e textos
#### Mas muitas listas sáo misturadas nao sendo o ideal

```python
produtos = ['tv', 'celular', 'mouse', 'teclado', 'tablet']
print(len(produtos))
```

Lista de Unidades Vendidas de cada Produto da Loja

```python
vendas = [1000, 1500, 350, 270, 900]


mais_vendido = max(vendas)
menor_venda = min(vendas)

index_mais_vendido = vendas.index(mais_vendido)
index_menor_venda = vendas.index(menor_venda)

print(f"O item mais vendido foi {produtos[index_mais_vendido]} com {mais_vendido} vendas.E o menos vendido foi o {produtos[index_menor_venda]} com {menor_venda} vendas.")
```
