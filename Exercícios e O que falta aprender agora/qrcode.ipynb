## QR Code com imagem

````python
import qrcode
from qrcode.constants import ERROR_CORRECT_H
from qrcode.image.styledpil import StyledPilImage

qr = qrcode.QRCode(error_correction=ERROR_CORRECT_H, box_size=8,border=2)

qr.add_data("https://www.linkedin.com/in/rayron-soares")
qr.make(fit=True)

img = qr.make_image(image_factory=StyledPilImage, embedded_image_path="1234.png")

img.save("qrcode_linkedin_styled.png")
````
