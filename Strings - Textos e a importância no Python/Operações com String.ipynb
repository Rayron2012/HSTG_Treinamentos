# Operações com String

str -> transforma número em string<br>
in -> verifica se um texto está contido dentro do outro<br>
operador + -> concatenar string<br>
format e {} -> substitui valores<br>
%s -> substitui textos<br>
%d -> substitui números decimais<br>

Vamos deixar uma cartilha para download

```python
faturamento = 1000
custo = 500
lucro = faturamento - custo
```

Uso do str() e do concatenar com +

```python
print ('O faturamento da loja foi de: ' + str(faturamento))
```

Uso do Format

```python
print('O faturamento foi de: {} '.format(faturamento))   

# É possível repetir as informações dentro das chaves uma ou mais vezes, lembrando apenas de copolocar o indice correto que esta dentro do format
# Uma vez que um indicce é atribuido, todos os outros devem ser atribuidos também, conforme abaixo
print('O faturamento foi de: {0} e o custo foi de: {1}. O lucro foi de: {2}. O faturamento novamente: {0}'.format(faturamento, custo, lucro))
```

Uso do %s e %d

```python
print ('O faturamento foi de: %d' % faturamento)
print ('O custo foi de: %s' % "1000")

# %d subistui um valor inteiro (numero)
# %s substitui uma string
```

Uso do in (mais exercícios práticos nas próximas aulas)

```python
print('@' in 'lira@gmail.com')
print('@' in 'lira.gmail.com')

# é muito usado para validar emails, mas não é isso
```
