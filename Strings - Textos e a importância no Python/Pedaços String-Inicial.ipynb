### Forma Básica

```python
precos = "Jan: 25, Fev: 27, Mar: 29"

# Primeiro caractere é inclusivo, o último é exclusivo

preco_jan = precos[0:7]
preco_fev = precos[8:12]
precos_mar = precos[13:]  
print(preco_jan, preco_fev, precos_mar)
```

### Posição Inicial e Final

```python
#  Para pegar apoenas os numeros dos precos pode ser usado o indice negativo
preco_jan_num = precos[-2:]
print(preco_jan_num)

preco_fev_num = precos[-11:-9]
print(preco_fev_num)

precos_mar_num = precos[-20:-18]
print(precos_mar_num)
```

### Posição Inicial e Final com Step

```python
codigo = "1.2.3.4,5,1,2.3.4,7.9"

numpos = codigo[::2]
numneg = codigo[::-2]
print(numpos)
print(numneg)

# Step pode ser usado para pular caracteres, aparentemente não é muito usado, pose ser util em casos para pegar caracteres em posições pares ou ímpares
```
